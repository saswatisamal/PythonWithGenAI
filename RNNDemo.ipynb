```python
import tensorflow as tf
from  tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN,Dense
import numpy as np

def loadData(text_file):
    with open(text_file,'r' ,encoding='utf-8') as f:
        text = f.read()
    return text
file_path="hp1.txt"
#loaded data from the file
text =loadData(file_path).lower()

#Pre-processing:
#Tokenize data word by word :
from tensorflow.keras.preprocessing.text import Tokenizer
#oov is out of vocab
tokenizer = Tokenizer(oov_token="<oov>")
tokenizer.fit_on_texts([text])

#Each unique word is added a unique identifier
word_index = tokenizer.word_index
print(word_index)

total_words = len(tokenizer.word_index) + 1
print(total_words)

#Convert every word to a token
print(tokenizer.texts_to_sequences([text])[0])

#










```
