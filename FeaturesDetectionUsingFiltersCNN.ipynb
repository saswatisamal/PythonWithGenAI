```python

#Vertical edges detection using the vertical edge detection filter
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
#converting to gray-scale
image=Image.open('/content/cat-image.png').convert("L")
#Resizing the image for simplicity
image=image.resize((256,256))
#Converting the data to float32 and normalize
image=np.array(image,dtype=np.float32)/255.0

 #Add batch and channel dimensions (1, 256, 256, 1)
 #For grayScale outputchannel=1
image=np.expand_dims(image,axis=(0,-1))
#Define vertical edge detection filter
vertical_filter = np.array([[1, 0, -1],
                            [1, 0, -1],
                            [1, 0, -1]], dtype=np.float32)
#Define horizontal edge detection filter

horizontal_filter = np.array([[1, 1,  1],
                              [0,  0,  0],
                              [-1,-1,-1]], dtype=np.float32)


# Reshape filters for TensorFlow (H, W, in_channels, out_channels)
vertical_filter = vertical_filter.reshape(3, 3, 1, 1)

horizontal_filter = horizontal_filter.reshape(3, 3, 1, 1)


#The padding available are SAME and VALID
#In Scipy.signal modes SAME ,VALID,FULL are present
vertical_edges=tf.nn.conv2d(image,vertical_filter,strides=[1,1,1,1],padding='SAME')

horizontal_edges=tf.nn.conv2d(image,horizontal_filter,strides=[1,1,1,1],padding='SAME')


# Plot original and filtered images
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(image.squeeze(), cmap="gray")
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(np.squeeze(vertical_edges), cmap='gray')
plt.title('Vertical Edges')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(np.squeeze(horizontal_edges), cmap='gray')
plt.title('Horizontal Edges')
plt.axis('off')


plt.show()














```

```python
#Blurr filter
blurr_filter = np.ones((5,5),dtype=np.float32)/25.0

blurr_filter=blurr_filter.reshape(5,5,1,1)

blurred_image=tf.nn.conv2d(image,blurr_filter,strides=[1,1,1,1],padding='SAME')


# Convert tensor to numpy for visualization
blurred_image = blurred_image.numpy().squeeze()

# Plot original and blurred images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image.squeeze(), cmap="gray")
plt.title("Original Image")
plt.axis("off")

#Blurred image
plt.subplot(1,2,2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')





```

```python
#Blurring the cat image even more :

#Blurr filter
blurr_filter = np.ones((10,10),dtype=np.float32)/100.0

blurr_filter=blurr_filter.reshape(10,10,1,1)

blurred_image=tf.nn.conv2d(image,blurr_filter,strides=[1,1,1,1],padding='SAME')


# Convert tensor to numpy for visualization
blurred_image = blurred_image.numpy().squeeze()

# Plot original and blurred images
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(image.squeeze(), cmap="gray")
plt.title("Original Image")
plt.axis("off")

#Blurred image
plt.subplot(1,2,2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')


```
